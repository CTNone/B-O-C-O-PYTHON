# react theo ngày

```python
# Xem xét dữ liệu trong cột 'reactions'
post_df_full1['reactions'] = post_df_full1['reactions'].apply(lambda x: dict(eval(x)) if isinstance(x, str) else x)


# post_df_full1['reactions'] = post_df_full1['reactions'].apply(lambda x : dict(eval(x)) )
post_df_full_reactions = post_df_full1['reactions'].apply(pd.Series )
```

```python
# gộp các cột lại vào dataframe
post_df_full_with_reactions = pd.concat([post_df_full1, post_df_full_reactions], axis=1).drop('reactions', axis=1)
post_df_full_with_reactions[['post_text','like','love','haha','wow','care','sad','angry',
                                 'shares','comments','reaction_count']]
```

```python
df = post_df_full_with_reactions

# Tìm chỉ số (index) của bài viết có lượng tương tác lớn nhất
max_interaction_index = df['reaction_count'].idxmax()

# Truy cập nội dung bài viết tương ứng với chỉ mục lớn nhất
max_interaction_post = df.loc[max_interaction_index, 'post_text']
LuongTT= df.loc[max_interaction_index, 'reaction_count']

# In nội dung bài viết có lượng tương tác lớn nhất
print("Bài viết thứ: " +str(max_interaction_index +1 ), ". Với tổng số lượng tương tác lớn nhất là: " + str(LuongTT) )
print( "Với nội dụng bài viết sau:")
print()
print(max_interaction_post)

max_interaction_index = df['shares'].idxmax()
print("số lượng share")
print(max_interaction_index +1)
print(df.loc[max_interaction_index, 'shares'] )
print( df.loc[max_interaction_index, 'post_text'] )

max_interaction_index = df['comments'].idxmax()
print()
print("số lượng coment")
print(max_interaction_index +1)
print(df.loc[max_interaction_index, 'comments'] )
```

```python
for i in range(0, len(post_df_full_with_reactions['like'].values), 1):
    if pd.isna(post_df_full_with_reactions['like'].values[i]):
        post_df_full_with_reactions['like'].values[i] = 0

    if pd.isna(post_df_full_with_reactions['love'].values[i]):
        post_df_full_with_reactions['love'].values[i] = 0

    if pd.isna(post_df_full_with_reactions['haha'].values[i]):
        post_df_full_with_reactions['haha'].values[i] = 0

    if pd.isna(post_df_full_with_reactions['wow'].values[i]):
        post_df_full_with_reactions['wow'].values[i] = 0

    if pd.isna(post_df_full_with_reactions['sad'].values[i]):
        post_df_full_with_reactions['sad'].values[i] = 0

    if pd.isna(post_df_full_with_reactions['angry'].values[i]):
        post_df_full_with_reactions['angry'].values[i] = 0

    if pd.isna(post_df_full_with_reactions['care'].values[i]):
        post_df_full_with_reactions['care'].values[i] = 0

    if pd.isna(post_df_full_with_reactions['shares'].values[i]):
        post_df_full_with_reactions['shares'].values[i] = 0

    if pd.isna(post_df_full_with_reactions['comments'].values[i]):
        post_df_full_with_reactions['comments'].values[i] = 0


post_df_full_with_reactions[['post_text','like','love','haha','wow','care','sad','angry',
                                 'shares','comments','reaction_count']]
```

```python

import matplotlib.pyplot as plt

# Giả sử df là DataFrame chứa các dữ liệu của bạn
df = post_df_full_with_reactions

total_reactions = df[['like', 'love', 'haha', 'wow', 'sad', 'angry', 'care']].sum(axis=1)
percentages = df[['like', 'love', 'haha', 'wow', 'sad', 'angry', 'care']].div(total_reactions, axis=0) * 100

# Tạo biểu đồ tròn
labels = ['Like', 'Love', 'Ha Ha', 'Wow', 'Sad', 'Angry', 'Care']
plt.figure(figsize=(10, 10))
patches, texts, autotexts = plt.pie(percentages.mean(), labels=None, autopct='', startangle=90)

# Tạo chú thích
legend_labels = [f"{label}: {percent:.1f}%" for label, percent in zip(labels, percentages.mean())]
plt.legend(patches, legend_labels, title='Reactions', loc='center left', bbox_to_anchor=(1, 0.5))

# Thiết lập tiêu đề
plt.title('Biểu Đồ Tỉ Lệ Phần Trăm Cảm Xúc')

# Hiển thị biểu đồ
plt.show()
```
